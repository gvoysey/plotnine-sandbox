```python
from plotnine import ggplot, aes
import plotnine as p9
import numpy as np 
import pandas as pd
```

```python
x = np.linspace(0, 2*np.pi, num=1000)
y = np.sin(x)
df = pd.DataFrame({'x':x, 'y':y})
df.head()
```

```python
(ggplot(df)
 + aes(x=df.x, y=df.y)
 + p9.geom_line()
 + p9.xlab("time")
 + p9.ylab("amplitude")
 + p9.ggtitle("a sine wave")
)
```
