```python
from plotnine import ggplot, aes
import plotnine as p9
import numpy as np 
import pandas as pd
```

```python
x = np.linspace(0, 2*np.pi, num=1000)
y = np.sin(x)
df = pd.DataFrame({'x':x, 'y':y})
df.head()
```

```python
p = (ggplot(df)
 + aes(x=x, y=y)
 + p9.geom_area())
```

```python
p.draw()
```
